## Experimental matplotlib backend for Robbie Remote Functions

Once your job completes, download the results from the [Robbie Portal](https://robbie.run/portal/app/my-runs) and view the image generated by matplotlib.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from robbie import remote
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
```

```python
@remote()
def run():
    plt.plot(x, y)
```

```python
run()
```
